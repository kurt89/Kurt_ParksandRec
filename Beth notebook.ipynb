```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
#Import Streaming dataframe from CSV
parks_df = pd.read_csv('parks.csv')
parks_df.head()
```

```python
population_df = pd.read_csv('Charlotte_Pop_and_Income.csv')
population_df.head()
```

```python
parksandpop_df = pd.merge(parks_df, population_df, on="zipcode")
parksandpop_df
```

```python
#Delete columns and confirm order of remaining columns
del parksandpop_df['STREETDIR']
del parksandpop_df['STREETSUFF']
del parksandpop_df['UPSIZE_TS']
del parksandpop_df['USERID']
del parksandpop_df['STREETTYPE']
del parksandpop_df['city']
parksandpop_df
```

```python
#Delete columns and confirm order of remaining columns
del parksandpop_df['PRKSIZE']
del parksandpop_df['PRKXCOORD']
del parksandpop_df['PRKYCOORD']
del parksandpop_df['PRKAGE']
del parksandpop_df['PRKADAACC']
parksandpop_df.head()
```

```python
#output to CSV
parksandpop_df.to_csv("parksandpop.csv", index=False, header=True)
```

```python
parksandpop_df.head()
```

```python
#output to CSV
parksformat_df.to_csv("parksdataframe.csv", index=False, header=True)
```
